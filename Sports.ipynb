```python
import warnings
warnings.filterwarnings('ignore')
from tqdm import tqdm_notebook
 
df = pd.read_csv('lap_times.csv')
df['position1'] = df.position
for i in tqdm_notebook(range(len(df.driverId)-1)):
    if df.driverId[i] == df.driverId[i+1]:
        df.position1[i+1] = df.position[i+1] - df.position[i]
    else:
        df.position1[i+1] = 0
df1 = df[['driverId', 'lap', 'position1', 'raceId']]
```

```python
a = list(df1.driverId.value_counts().index)
```

```python
def maxim(df):
    m = df.position1.argmax()
    return [df.lap[m], df.position1[m]]
```

```python
res_driver = []
res_lap = []
res_diff = []
for i in a:
    df_t = df1[df1.driverId == i].reset_index()
    lap, diff = maxim(df_t)
    res_lap.append(lap)
    res_diff.append(diff)
    res_driver.append(i)
    
result = pd.DataFrame(data = {'driverId': res_driver, 'lap': res_lap, 'max_diff': res_diff})
result.sort_values('driverId')
```

```python
race_lap = {}
df = pd.read_csv('lap_times.csv')
a = list(df.raceId.value_counts().index)
for i in a:
    df_t = df[df.raceId == i].reset_index()
    m = df_t.lap.argmax()
    race_lap[i] = df_t.lap[m]
```

```python
count = 0
df = pd.read_csv('lap_times.csv')
df_clear = pd.DataFrame(columns = ['index','raceId','driverId','position'])
for i in tqdm_notebook(range(len(df))):
    if race_lap.get(df.raceId[i]) == df.lap[i] and df.position[i] < 11 :
        df_clear.loc[count] = [i,df.raceId[i],df.driverId[i],df.position[i]]  
        count+=1
df_clear
```

```python
df_clear = df_clear.astype('int64')
```
